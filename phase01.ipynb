```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import glob 
import matplotlib.font_manager as fm
print("  ✅ 1. 라이브러리 로딩 완료")

# 2. 한글 폰트 경로 직접 지정
print("  ▶ 2. 한글 폰트 설정을 시작합니다...")
try:
    font_path = r"C:\WINDOWS\Fonts\malgun.ttf"
    font_prop = fm.FontProperties(fname=font_path).get_name()
    plt.rc('font', family=font_prop); plt.rcParams['axes.unicode_minus'] = False 
    sns.set_style('whitegrid'); print(f"    ✅ 한글 폰트 '{font_prop}' 설정을 완벽하게 적용했습니다!")
except Exception as e: print(f"    ❌ 폰트 설정 오류: {e}")
    
# 3. 모든 CSV 파일이 있는 폴더 경로 지정
folder_path = './data/전국오염원 조사자료'
all_csv_files = glob.glob(os.path.join(folder_path, '*.csv'))
print(f"  ✅ 3. '{folder_path}'에서 총 {len(all_csv_files)}개의 CSV 파일을 발견했습니다.")


final_feature_recommendations = {}
print("\n--- 모든 준비 완료. 이제 당신의 모든 EDA를 통합하여 심층 분석을 시작합니다. ---\n")

# [분석 단계]
for file_path in all_csv_files:
    file_name = os.path.basename(file_path)
    # 분석 결과물인 feature_matrix 파일들은 건너뜁니다.
    if 'feature_matrix' in file_name:
        print(f"\n\n==================== [ 확인: {file_name} ] ====================")
        print("  ▶ 이 파일은 분석 결과물이므로 건너뜁니다.")
        continue
        
    print(f"\n\n==================== [ 심층 분석 시작: {file_name} ] ====================")
    try:
        df = pd.read_csv(file_path, encoding='utf-8', dtype=str, low_memory=False)
        print(f"  ▶ 파일 로딩 성공 (총 {len(df)}개 행)")
        
        
        # [A] '방류량' 파일
        if '방류량' in file_name:
            print("  ▶ [핵심 분석] 실질적 오염 기여도(총 오염 부하량) TOP 10 분석")
            numeric_cols = [col for col in df.columns if any(key in col for key in ['유량', 'BOD', 'COD', 'SS', 'TN', 'TP'])];
            for col in numeric_cols: df[col] = pd.to_numeric(df[col], errors='coerce')
            flow_col = [col for col in df.columns if '유량(㎥/일)' in col][0]; poll_cols = [col for col in df.columns if any(key in col for key in ['BOD(mg/ℓ)', 'COD(mg/ℓ)', 'SS(mg/ℓ)', 'TN(mg/ℓ)', 'TP(mg/ℓ)'])]
            for col in poll_cols: df[f'{col}_부하량'] = (df[flow_col] * df[col]) / 1000
            load_cols = [f'{col}_부하량' for col in poll_cols]; df['총오염부하량'] = df[load_cols].sum(axis=1, skipna=True)
            top10_data = df.groupby('처리시설명')['총오염부하량'].sum().nlargest(10)
            plt.figure(figsize=(12, 6)); sns.barplot(x=top10_data.values, y=top10_data.index, palette='YlOrRd_r'); plt.title(f'[{file_name}] 실질적 오염 기여도 TOP 10', fontproperties=font_prop, fontsize=16); plt.xlabel("총 오염 부하량 (kg/일)", fontproperties=font_prop); plt.ylabel('처리시설명', fontproperties=font_prop); plt.xticks(fontproperties=font_prop); plt.yticks(fontproperties=font_prop); plt.show()
            print("  💡 상세 동적 인사이트:"); [print(f"       {i+1}위: {item} (하루 약 {val:,.1f} kg)") for i, (item, val) in enumerate(top10_data.items())]
            final_feature_recommendations['환경기초시설_총방류부하량'] = file_name
            print(f"     ▶▶ [설계도 추가] '환경기초시설_총방류부하량' 컬럼을 추가합니다.")
        
        # [B] '기본현황' 파일
        elif '기본현황' in file_name:
            print("\n  ▶ [심층 분석 1] 시설 밀집도 TOP 10:"); top10_data = df['시군구'].value_counts().nlargest(10)
            plt.figure(figsize=(12, 6)); sns.barplot(x=top10_data.values, y=top10_data.index, palette='Greens_d'); plt.title(f'[{file_name}] 처리시설 밀집도 TOP 10', fontproperties=font_prop, fontsize=16); plt.xlabel("시설 개수", fontproperties=font_prop); plt.ylabel('시군구', fontproperties=font_prop); plt.xticks(fontproperties=font_prop); plt.yticks(fontproperties=font_prop); plt.show()
            print("  💡 상세 동적 인사이트:"); [print(f"       {i+1}위: {item} ({val} 개)") for i, (item, val) in enumerate(top10_data.items())]
            final_feature_recommendations['환경기초시설_시설밀집도'] = file_name
            print(f"     ▶▶ [설계도 추가] '환경기초시설_시설밀집도' 컬럼을 추가합니다.")
            
            print("\n  ▶ [심층 분석 2] 처리용량과 가동율의 상관관계 분석:")
            df['처리용량'] = pd.to_numeric(df['처리용량'], errors='coerce'); df['가동율(%)'] = pd.to_numeric(df['가동율(%)'], errors='coerce')
            subset_df = df[['처리용량', '가동율(%)']].dropna()
            if len(subset_df) > 1:
                correlation = subset_df.corr(numeric_only=True).iloc[0, 1]
                plt.figure(figsize=(10, 6)); sns.regplot(data=subset_df, x='처리용량', y='가동율(%)', scatter_kws={'alpha':0.2}); plt.title(f'[{file_name}] 처리용량과 가동율의 상관관계', fontproperties=font_prop, fontsize=15); plt.xlabel("처리용량", fontproperties=font_prop); plt.ylabel("가동율(%)", fontproperties=font_prop); plt.xticks(fontproperties=font_prop); plt.yticks(fontproperties=font_prop); plt.show()
                print(f"  💡 동적 인사이트: 처리용량과 가동율의 상관계수는 {correlation:.4f}로, 시설 규모와 운영 효율성은 거의 관계가 없음을 의미합니다. 이는 AI 특성에서 제외하거나 낮은 가중치를 부여할 근거가 됩니다.")
        
        # [C] '총유입량', '관로이송량', '직접이송량' 파일 (시계열 분석)
        elif any(keyword in file_name for keyword in ['총유입량', '관로이송량', '직접이송량']):
            print(f"  ▶ [심층 분석] '{file_name}'의 30일 이동평균 유량 추세 분석")
            df['운영일자'] = pd.to_datetime(df['운영일자'], errors='coerce')
            flow_col_name = next((col for col in df.columns if '유량' in col and ('(㎥/일)' in col or '(m^3/일)' in col)), None)
            if flow_col_name:
                df[flow_col_name] = pd.to_numeric(df[flow_col_name], errors='coerce')
                daily_sum = df.groupby('운영일자')[flow_col_name].sum()
                rolling_avg_30d = daily_sum.rolling(window=30).mean()
                plt.figure(figsize=(14, 6)); rolling_avg_30d.plot(); plt.title(f'[{file_name}] 30일 이동평균 유량 추세', fontproperties=font_prop, fontsize=16); plt.xlabel("운영일자", fontproperties=font_prop); plt.ylabel("30일 평균 유량", fontproperties=font_prop); plt.xticks(fontproperties=font_prop); plt.yticks(fontproperties=font_prop); plt.show()
                if not rolling_avg_30d.empty and rolling_avg_30d.notna().any():
                    peak_date = rolling_avg_30d.idxmax(); peak_value = rolling_avg_30d.max()
                    print(f"  💡 동적 인사이트: 30일 평균 유량은 '{peak_date.strftime('%Y-%m-%d')}'에 약 {peak_value:,.0f} (㎥/일)로 최대치를 기록했습니다. 이는 AI의 이상치 탐지 모델 학습에 활용될 수 있습니다.")
        
        # [D] '생활계_인구' 또는 '생활계_물사용량' 파일
        elif '생활계_인구' in file_name:
            print("  ▶ [심층 분석] 생활계 '인구'와 '물사용량' 데이터를 통합하여, '미처리 하수 배출량' TOP 10 분석")
            df_water_path = os.path.join(folder_path, '1_2__생활계_물사용량_.csv')
            if os.path.exists(df_water_path):
                df_water = pd.read_csv(df_water_path, encoding='utf-8', dtype=str, low_memory=False)
                df_water['지역'] = df_water['시도'].str.strip() + '_' + df_water['시군구'].str.strip()
                water_cols = [col for col in df_water.columns if '물사용량' in col and '하수미처리' in col]
                for col in water_cols: df_water[col] = pd.to_numeric(df_water[col], errors='coerce')
                df_water['미처리하수배출량'] = df_water[water_cols].sum(axis=1)
                top10_data = df_water.groupby('지역')['미처리하수배출량'].sum().nlargest(10)
                plt.figure(figsize=(12, 6)); sns.barplot(x=top10_data.values, y=top10_data.index, palette='cool'); plt.title(f'[{file_name}] 미처리 하수 배출량 TOP 10', fontproperties=font_prop, fontsize=16); plt.xlabel("미처리 하수 배출량(m³/일)", fontproperties=font_prop); plt.ylabel('지역', fontproperties=font_prop); plt.xticks(fontproperties=font_prop); plt.yticks(fontproperties=font_prop); plt.show()
                print("  💡 상세 동적 인사이트:"); [print(f"       {i+1}위: {item} (하루 약 {val:,.0f} m³)") for i, (item, val) in enumerate(top10_data.items())]
                final_feature_recommendations['생활계_미처리하수배출량'] = '1_1__생활계_인구_.csv & 1_2__생활계_물사용량_.csv'
                print(f"     ▶▶ [설계도 추가] '생활계_미처리하수배출량' 컬럼을 추가합니다.")
            else:
                print("    - 경고: '1_2__생활계_물사용량_.csv' 파일을 찾을 수 없어 통합 분석을 건너뜁니다.")

        # [E] '축산계' 파일
        elif '축산계' in file_name:
            print("  ▶ [심층 분석] '고위험 배출량(미처리+정화방류)' TOP 10 분석")
            df['지역'] = df['시도'].str.strip() + '_' + df['시군구'].str.strip()
            high_risk_cols = [col for col in df.columns if '미처리' in col or '정화방류' in col]
            for col in high_risk_cols: df[col] = pd.to_numeric(df[col], errors='coerce')
            df['축산계_고위험배출량'] = df[high_risk_cols].sum(axis=1)
            top10_data = df.groupby('지역')['축산계_고위험배출량'].sum().nlargest(10)
            plt.figure(figsize=(12, 6)); sns.barplot(x=top10_data.values, y=top10_data.index, palette='autumn'); plt.title(f'[{file_name}] 축산계 고위험 배출량 TOP 10', fontproperties=font_prop, fontsize=16); plt.xlabel("미처리+정화방류 분뇨량(톤/일)", fontproperties=font_prop); plt.ylabel('지역', fontproperties=font_prop); plt.xticks(fontproperties=font_prop); plt.yticks(fontproperties=font_prop); plt.show()
            print("  💡 상세 동적 인사이트:"); [print(f"       {i+1}위: {item} (하루 약 {val:,.1f} 톤)") for i, (item, val) in enumerate(top10_data.items())]
            final_feature_recommendations['축산계_고위험배출량'] = file_name
            print(f"     ▶▶ [설계도 추가] '축산계_고위험배출량' 컬럼을 추가합니다.")
        
        # [F] '산업계' 파일
        elif '산업계' in file_name:
            print("  ▶ [심층 분석] '총계_폐수방류량' TOP 10 분석")
            df['지역'] = df['시도'].str.strip() + '_' + df['시군구'].str.strip(); target_col = '총계_폐수방류량(㎥/일)'
            df[target_col] = pd.to_numeric(df[target_col], errors='coerce')
            top10_data = df.groupby('지역')[target_col].sum().nlargest(10)
            plt.figure(figsize=(12, 6)); sns.barplot(x=top10_data.values, y=top10_data.index, palette='Purples_d'); plt.title(f'[{file_name}] 산업계 총 폐수방류량 TOP 10', fontproperties=font_prop, fontsize=16); plt.xlabel("총 폐수방류량(㎥/일)", fontproperties=font_prop); plt.ylabel('지역', fontproperties=font_prop); plt.xticks(fontproperties=font_prop); plt.yticks(fontproperties=font_prop); plt.show()
            print("  💡 상세 동적 인사이트:"); [print(f"       {i+1}위: {item} (하루 약 {val:,.0f} m³)") for i, (item, val) in enumerate(top10_data.items())]
            final_feature_recommendations['산업계_총폐수방류량'] = file_name
            print(f"     ▶▶ [설계도 추가] '산업계_총폐수방류량' 컬럼을 추가합니다.")
        
        # [G] '토지계' 파일
        elif '토지계' in file_name:
            print("  ▶ [심층 분석] 도시/산업 지역 면적(불투수층) TOP 10 분석")
            df['지역'] = df['시도'].str.strip() + '_' + df['시군구'].str.strip(); impervious_cols = ['대지', '공장용지', '도로', '주차장']
            impervious_cols_found = [col for col in df.columns if any(ic in col for ic in impervious_cols)]
            for col in impervious_cols_found: df[col] = pd.to_numeric(df[col], errors='coerce')
            df['토지계_불투수층면적'] = df[impervious_cols_found].sum(axis=1)
            top10_data = df.groupby('지역')['토지계_불투수층면적'].sum().nlargest(10)
            plt.figure(figsize=(12, 6)); sns.barplot(x=top10_data.values, y=top10_data.index, palette='bone_r'); plt.title(f'[{file_name}] 불투수층 면적 TOP 10', fontproperties=font_prop, fontsize=16); plt.xlabel("불투수층 면적(㎡)", fontproperties=font_prop); plt.ylabel('지역', fontproperties=font_prop); plt.xticks(fontproperties=font_prop); plt.yticks(fontproperties=font_prop); plt.show()
            print("  💡 상세 동적 인사이트:"); [print(f"       {i+1}위: {item} (약 {val:,.0f} ㎡)") for i, (item, val) in enumerate(top10_data.items())]
            final_feature_recommendations['토지계_불투수층면적'] = file_name
            print(f"     ▶▶ [설계도 추가] '토지계_불투수층면적' 컬럼을 추가합니다.")

        # [H] '양식계' 파일
        elif '양식계' in file_name:
            print("  ▶ [심층 분석] '총 사료사용량' TOP 10 분석")
            df['지역'] = df['시도'].str.strip() + '_' + df['시군구'].str.strip()
            feed_cols = [col for col in df.columns if '사료사용량' in col]
            for col in feed_cols: df[col] = pd.to_numeric(df[col], errors='coerce')
            df['총사료사용량'] = df[feed_cols].sum(axis=1)
            top10_data = df.groupby('지역')['총사료사용량'].sum().nlargest(10)
            plt.figure(figsize=(12, 6)); sns.barplot(x=top10_data.values, y=top10_data.index, palette='coolwarm'); plt.title(f'[{file_name}] 양식계 총 사료사용량 TOP 10', fontproperties=font_prop, fontsize=16); plt.xlabel("총 사료사용량(kg/월)", fontproperties=font_prop); plt.ylabel('지역', fontproperties=font_prop); plt.xticks(fontproperties=font_prop); plt.yticks(fontproperties=font_prop); plt.show()
            print("  💡 상세 동적 인사이트:"); [print(f"       {i+1}위: {item} (월 약 {val:,.0f} kg)") for i, (item, val) in enumerate(top10_data.items())]
            final_feature_recommendations['양식계_총사료사용량'] = file_name
            print(f"     ▶▶ [설계도 추가] '양식계_총사료사용량' 컬럼을 추가합니다.")
            
        # [I] '매립계' 파일
        elif '매립계' in file_name:
            print("  ▶ [심층 분석] 침출수 '유량 처리율'이 가장 낮은 시설 TOP 10 (누출/증발 의심)")
            gen_flow_col = [col for col in df.columns if '발생' in col and '유량' in col][0]; dis_flow_col = [col for col in df.columns if '방류' in col and '유량' in col][0]
            for col in [gen_flow_col, dis_flow_col]: df[col] = pd.to_numeric(df[col], errors='coerce')
            df['처리율'] = df[dis_flow_col] / df[gen_flow_col]
            top10_data = df.groupby('처리시설명')['처리율'].mean().dropna().nsmallest(10)
            plt.figure(figsize=(12, 6)); top10_data.plot(kind='bar', color='tomato'); plt.title(f'[{file_name}] 침출수 유량 처리율 최하위 10개 시설', fontproperties=font_prop, fontsize=16); plt.xlabel("처리시설명", fontproperties=font_prop); plt.ylabel("평균 처리율", fontproperties=font_prop); plt.xticks(fontproperties=font_prop); plt.yticks(fontproperties=font_prop); plt.show()
            print("  💡 상세 동적 인사이트: 아래 시설들은 발생한 침출수 대비 방류량이 현저히 적어, 처리 비효율 또는 중간 누출/증발 가능성을 시사합니다."); [print(f"       - {item} (평균 처리율: {val:.2%})") for i, (item, val) in enumerate(top10_data.items())]
            final_feature_recommendations['매립계_평균처리율'] = file_name
            print(f"     ▶▶ [설계도 추가] '매립계_평균처리율' 컬럼을 추가합니다.")

        # [J] '기타수질오염원' 파일
        elif '기타수질오염원' in file_name:
            print("  ▶ [심층 분석] 기타 수질오염원 총 개수가 가장 많은 지역 TOP 10")
            df['지역'] = df['시도'].str.strip() + '_' + df['시군구'].str.strip(); target_col = '총계(개소)'
            df[target_col] = pd.to_numeric(df[target_col], errors='coerce')
            top10_data = df.groupby('지역')[target_col].sum().nlargest(10)
            plt.figure(figsize=(12, 6)); sns.barplot(x=top10_data.values, y=top10_data.index, palette='viridis'); plt.title(f'[{file_name}] 기타수질오염원 총 개수 TOP 10', fontproperties=font_prop, fontsize=16); plt.xlabel("총 개소", fontproperties=font_prop); plt.ylabel('지역', fontproperties=font_prop); plt.xticks(fontproperties=font_prop); plt.yticks(fontproperties=font_prop); plt.show()
            print("  💡 상세 동적 인사이트:"); [print(f"       {i+1}위: {item} ({val} 개)") for i, (item, val) in enumerate(top10_data.items())]
            final_feature_recommendations['기타수질오염원_총개소'] = file_name
            print(f"     ▶▶ [설계도 추가] '기타수질오염원_총개소' 컬럼을 추가합니다.")
        
        # 이전에 심층 분석한 '생활계_물사용량' 파일은 여기서 별도로 처리하지 않도록 함
        elif '생활계_물사용량' in file_name:
             print(f"  ▶ 이 파일({file_name})은 '생활계_인구' 파일과 통합하여 분석되었으므로, 여기서는 넘어갑니다.")
             
        else:
            print(f"  ▶ 이 파일({file_name})은 현재 정의된 심층 분석 대상이 아니므로, 데이터 구조만 확인합니다:\n{df.head()}")

    except Exception as e:
        print(f"  ❌ 분석 중 심각한 오류가 발생했습니다: {e}")


if not final_feature_recommendations: print("추천할 만한 결합 컬럼을 찾지 못했습니다.")
else:
    print("---------- [ 1. 각 핵심 컬럼의 출처 ] ----------\n")
    for feature, source_csv in sorted(final_feature_recommendations.items()): print(f"▶ **결합 필요 컬럼: {feature}**\n  *(.CSV 원본 파일: {source_csv})*\n")
    final_columns_list = sorted(list(final_feature_recommendations.keys()))
    print("\n---------- [ 2. 최종 결론 및 다음 단계 ] ----------\n")
    print("\n**▶▶ 최종적으로 다음의 컬럼들을 '시도_시군구' 기준으로 결합하는 것을 제안합니다:**"); print(final_columns_list)
print("\n==================== 모든 통합 분석 및 설계 완료 ====================")
```

```python
print("▶ [준비 단계] 라이브러리 로딩 및 설정")
import pandas as pd
import os
import glob
from functools import reduce
from sklearn.preprocessing import MinMaxScaler
from thefuzz import process
from tqdm import tqdm

folder_path = './data/전국오염원 조사자료'
all_csv_files = glob.glob(os.path.join(folder_path, '*.csv'))
print(f"  ✅ '{folder_path}'에서 총 {len(all_csv_files)}개의 CSV 파일을 발견했습니다.")

# ------------------------------------------------------------------------------
# [사전 작업] 위치 정보 지도 생성
# ------------------------------------------------------------------------------
print("  ▶ [사전 작업] 처리시설별 위치 정보(지역)를 생성합니다.")
location_map = {}
clean_name_list = []
try:
    basic_info_path = os.path.join(folder_path, '환경기초시설_1__기본현황.csv')
    df_basic = pd.read_csv(basic_info_path, encoding='utf-8', dtype=str, low_memory=False)
    df_basic['지역'] = df_basic['시도'].str.strip() + '_' + df_basic['시군구'].str.strip()
    df_basic['처리시설명'] = df_basic['처리시설명'].str.strip()
    location_map = pd.Series(df_basic['지역'].values, index=df_basic['처리시설명']).to_dict()
    clean_name_list = df_basic['처리시설명'].unique().tolist()
    print("    ✅ 처리시설명-지역 매핑 및 유사도 비교 기준 목록 생성 완료.")
except Exception as e:
    print(f"    ❌ 치명적 오류: '기본현황' 파일 처리 중 문제가 발생했습니다: {e}")

regional_summary_list = []
print("  ✅ 설계를 위한 데이터 취합 준비 완료.")
print("\n--- 지금부터 각 CSV 파일에서 핵심 지표를 추출하여 통합 데이터 생성을 시작합니다. ---\n")

# ------------------------------------------------------------------------------
# [핵심 로직] "선 요약, 후 적용" 방식으로 '방류량' 파일 초고속 처리
# ------------------------------------------------------------------------------
print("==================== [ '방류량' 파일 통합 및 지능형 매칭 처리 시작 ] ====================")
discharge_files = [f for f in all_csv_files if '방류량' in os.path.basename(f)]
if discharge_files and location_map:
    df_discharge_total = pd.concat([pd.read_csv(f, encoding='utf-8', dtype=str, low_memory=False) for f in discharge_files], ignore_index=True)
    print(f"  ▶ {len(discharge_files)}개의 '방류량' 파일을 하나로 통합했습니다. (총 {len(df_discharge_total)} 행)")

    unique_dirty_names = df_discharge_total['처리시설명'].str.strip().dropna().unique()
    
    # ★가장 중요한 수정 1★: 수술 전 상태 보고
    initial_missing_count = sum(1 for name in unique_dirty_names if name not in location_map)
    print(f"  ▶ 수술 전 상태: {initial_missing_count}개의 처리시설명이 일치하지 않아 데이터 누락이 예상됩니다.")
    print(f"  ▶ {len(unique_dirty_names)}개의 고유 처리시설명에 대한 '오타 교정 지도' 생성을 시작합니다.")

    fuzzy_match_map = {}
    for dirty_name in tqdm(unique_dirty_names, desc="  - 오타 교정 지도 생성 중"):
        if dirty_name in location_map:
            fuzzy_match_map[dirty_name] = dirty_name # 이미 이름이 같으면 바로 추가
        else:
            best_match = process.extractOne(dirty_name, clean_name_list)
            if best_match and best_match[1] >= 90:
                fuzzy_match_map[dirty_name] = best_match[0]
    print("    ✅ '오타 교정 지도' 생성 완료.")

    df_discharge_total['처리시설명_매칭'] = df_discharge_total['처리시설명'].str.strip().map(fuzzy_match_map)
    df_discharge_total['지역'] = df_discharge_total['처리시설명_매칭'].map(location_map)

    # ★가장 중요한 수정 2★: 수술 후 결과 보고
    final_missing_rows = df_discharge_total['지역'].isnull().sum()
    total_rows = len(df_discharge_total)
    saved_rows = total_rows - final_missing_rows
    
    print("\n---------- [ 데이터 복구 결과 보고 ] ----------")
    print(f"  - 총 데이터: {total_rows} 건")
    print(f"  - 최종 누락 데이터: {final_missing_rows} 건")
    print(f"  - 최종 복구 및 사용된 데이터: {saved_rows} 건 (전체의 {saved_rows/total_rows:.2%})")
    print("-------------------------------------------")

    if final_missing_rows > 0:
        df_discharge_total.dropna(subset=['지역'], inplace=True)

    # [이하 계산 로직은 동일]
    numeric_cols = [col for col in df_discharge_total.columns if any(key in col for key in ['유량', 'BOD', 'COD', 'SS', 'TN', 'TP'])]
    for col in numeric_cols: df_discharge_total[col] = pd.to_numeric(df_discharge_total[col], errors='coerce')
    flow_col = [col for col in df_discharge_total.columns if '유량(㎥/일)' in col][0]
    poll_cols = [col for col in df_discharge_total.columns if any(key in col for key in ['BOD(mg/ℓ)', 'COD(mg/ℓ)', 'SS(mg/ℓ)', 'TN(mg/ℓ)', 'TP(mg/ℓ)'])]
    for col in poll_cols: df_discharge_total[f'{col}_부하량'] = (df_discharge_total[flow_col] * df_discharge_total[col]) / 1000
    load_cols = [f'{col}_부하량' for col in poll_cols]
    df_discharge_total['환경기초시설_총방류부하량'] = df_discharge_total[load_cols].sum(axis=1, skipna=True)
    summary = df_discharge_total.groupby('지역')[['환경기초시설_총방류부하량']].sum()
    regional_summary_list.append(summary)
    print("    ✅ '환경기초시설_총방류부하량' 계산 완료.")

# ------------------------------------------------------------------------------
# [나머지 파일들 처리]
# ------------------------------------------------------------------------------
for file_path in all_csv_files:
    file_name = os.path.basename(file_path)
    if '방류량' in file_name or 'feature_matrix' in file_name:
        continue
    print(f"==================== [ 데이터 추출 중: {file_name} ] ====================")
    try:
        df = pd.read_csv(file_path, encoding='utf-8', dtype=str, low_memory=False)
        if '기본현황' in file_name:
            print("  ▶ '환경기초시설_시설밀집도' 계산 중..."); df['지역'] = df['시도'].str.strip() + '_' + df['시군구'].str.strip(); summary = df.groupby('지역').size().reset_index(name='환경기초시설_시설밀집도').set_index('지역'); regional_summary_list.append(summary); print("    ✅ 완료.")
        elif '생활계_인구' in file_name:
            print("  ▶ '생활계_미처리하수배출량' 계산 중..."); df_water_path = os.path.join(folder_path, '1_2__생활계_물사용량_.csv')
            if os.path.exists(df_water_path):
                df_water = pd.read_csv(df_water_path, encoding='utf-8', dtype=str, low_memory=False); df_water['지역'] = df_water['시도'].str.strip() + '_' + df_water['시군구'].str.strip(); water_cols = [col for col in df_water.columns if '물사용량' in col and '하수미처리' in col]; 
                for col in water_cols: df_water[col] = pd.to_numeric(df_water[col], errors='coerce')
                df_water['생활계_미처리하수배출량'] = df_water[water_cols].sum(axis=1); summary = df_water.groupby('지역')[['생활계_미처리하수배출량']].sum(); regional_summary_list.append(summary); print("    ✅ 완료.")
            else: print("    ❌ 경고: '1_2__생활계_물사용량_.csv' 파일을 찾을 수 없어 건너뜁니다.")
        elif '축산계' in file_name:
            print("  ▶ '축산계_고위험배출량' 계산 중..."); df['지역'] = df['시도'].str.strip() + '_' + df['시군구'].str.strip(); high_risk_cols = [col for col in df.columns if '미처리' in col or '정화방류' in col];
            for col in high_risk_cols: df[col] = pd.to_numeric(df[col], errors='coerce')
            df['축산계_고위험배출량'] = df[high_risk_cols].sum(axis=1); summary = df.groupby('지역')[['축산계_고위험배출량']].sum(); regional_summary_list.append(summary); print("    ✅ 완료.")
        elif '산업계' in file_name:
            print("  ▶ '산업계_총폐수방류량' 계산 중..."); df['지역'] = df['시도'].str.strip() + '_' + df['시군구'].str.strip(); target_col = '총계_폐수방류량(㎥/일)'; df[target_col] = pd.to_numeric(df[target_col], errors='coerce'); df.rename(columns={target_col: '산업계_총폐수방류량'}, inplace=True); summary = df.groupby('지역')[['산업계_총폐수방류량']].sum(); regional_summary_list.append(summary); print("    ✅ 완료.")
        elif '토지계' in file_name:
            print("  ▶ '토지계_불투수층면적' 계산 중..."); df['지역'] = df['시도'].str.strip() + '_' + df['시군구'].str.strip(); impervious_cols = ['대지', '공장용지', '도로', '주차장']; impervious_cols_found = [col for col in df.columns if any(ic in col for ic in impervious_cols)];
            for col in impervious_cols_found: df[col] = pd.to_numeric(df[col], errors='coerce')
            df['토지계_불투수층면적'] = df[impervious_cols_found].sum(axis=1); summary = df.groupby('지역')[['토지계_불투수층면적']].sum(); regional_summary_list.append(summary); print("    ✅ 완료.")
        elif '양식계' in file_name:
            print("  ▶ '양식계_총사료사용량' 계산 중..."); df['지역'] = df['시도'].str.strip() + '_' + df['시군구'].str.strip(); feed_cols = [col for col in df.columns if '사료사용량' in col];
            for col in feed_cols: df[col] = pd.to_numeric(df[col], errors='coerce')
            df['양식계_총사료사용량'] = df[feed_cols].sum(axis=1); summary = df.groupby('지역')[['양식계_총사료사용량']].sum(); regional_summary_list.append(summary); print("    ✅ 완료.")
        elif '매립계' in file_name:
            print("  ▶ '매립계_평균처리율' 계산 중..."); df['지역'] = df['시도'].str.strip() + '_' + df['시군구'].str.strip(); gen_flow_col = [col for col in df.columns if '발생' in col and '유량' in col][0]; dis_flow_col = [col for col in df.columns if '방류' in col and '유량' in col][0];
            for col in [gen_flow_col, dis_flow_col]: df[col] = pd.to_numeric(df[col], errors='coerce')
            df['처리율'] = df[dis_flow_col] / df[gen_flow_col]; df.rename(columns={'처리율': '매립계_평균처리율'}, inplace=True); summary = df.groupby('지역')[['매립계_평균처리율']].mean(); regional_summary_list.append(summary); print("    ✅ 완료.")
        elif '기타수질오염원' in file_name:
            print("  ▶ '기타수질오염원_총개소' 계산 중..."); df['지역'] = df['시도'].str.strip() + '_' + df['시군구'].str.strip(); target_col = '총계(개소)'; df[target_col] = pd.to_numeric(df[target_col], errors='coerce'); df.rename(columns={target_col: '기타수질오염원_총개소'}, inplace=True); summary = df.groupby('지역')[['기타수질오염원_총개소']].sum(); regional_summary_list.append(summary); print("    ✅ 완료.")
    except Exception as e:
        print(f"  ❌ 분석 중 오류 발생: {e}")

# ------------------------------------------------------------------------------
# [최종 정리 및 파일 생성 단계]
# ------------------------------------------------------------------------------
print("\n\n==================== [ 최종 통합 및 정규화 ] ====================")
if not regional_summary_list:
    print("  ❌ 처리된 데이터가 없어 파일을 생성할 수 없습니다.")
else:
    feature_matrix = reduce(lambda left, right: pd.merge(left, right, on='지역', how='outer'), regional_summary_list)
    feature_matrix.fillna(0, inplace=True)
    
    print("  ▶ 모든 핵심 지표를 성공적으로 병합했습니다.")
    print("  ▶ 최종 데이터 컬럼:", feature_matrix.columns.tolist())
    print("  ▶ 최종 데이터 Shape:", feature_matrix.shape)
    
    scaler = MinMaxScaler()
    feature_matrix_normalized_data = scaler.fit_transform(feature_matrix)
    feature_matrix_normalized = pd.DataFrame(feature_matrix_normalized_data, 
                                             index=feature_matrix.index, 
                                             columns=feature_matrix.columns)
    
    output_path = os.path.join(folder_path, "feature_matrix_by_sigungu.csv")
    feature_matrix_normalized.to_csv(output_path, encoding='utf-8-sig')
    
    print(f"  ✅ 최종 결과물이 '{output_path}' 경로에 저장되었습니다.")
    print("  - 데이터 복구 작업이 성공적으로 완료되었습니다.")
    print("================================================================")
```

### 수동 맵핑 버전 

```python
# ==============================================================================
# 0. 환경 설정 및 모든 라이브러리 임포트
# ==============================================================================
print("■■■ 0. 환경 설정 및 라이브러리 임포트 시작 ■■■")
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import seaborn as sns
import warnings
import re

warnings.filterwarnings('ignore')

# ==============================================================================
# 1. 안정적인 한글 폰트 설정 및 전문가용 시각화 테마 적용
# ==============================================================================
print("\n■■■ 1. 한글 폰트 및 시각화 테마 설정 시작 ■■■")
try:
    font_path = "c:/Windows/Fonts/malgun.ttf"
    font_name = fm.FontProperties(fname=font_path).get_name()
    # [업그레이드] Seaborn 테마를 설정하고, 기본 폰트를 지정합니다.
    sns.set_theme(style='whitegrid', font=font_name, palette='viridis')
    # 마이너스 기호 깨짐 방지
    plt.rcParams['axes.unicode_minus'] = False
    print("  ✅ 폰트 및 전문가용 시각화 테마 적용 완료.")
except Exception as e:
    print(f"  - 폰트 설정 중 오류 발생: {e}")

# ==============================================================================
# 2. 데이터 로드 및 사전 작업: GIMS 데이터에 한글 지역명 추가
# ==============================================================================
print("\n■■■ 2. 데이터 로드 및 핵심 사전 작업 시작 ■■■")

def load_csv(path):
    try: return pd.read_csv(path, encoding='cp949', low_memory=False)
    except UnicodeDecodeError: return pd.read_csv(path, encoding='utf-8', low_memory=False)

gims_df_orig = load_csv('./data/gims_master_data2.csv')
feature_df = load_csv('./data/feature_matrix_by_sigungu.csv')
standards_df = load_csv('./data/groundwater_quality_standards.csv')

if any(df is None for df in [gims_df_orig, feature_df, standards_df]):
    print("\n‼️ 일부 파일 로드에 실패하여 분석을 중단합니다."); exit()
print("  ✅ 모든 원본 파일 로드 성공.")

# gims_df를 복사하여 사용 (원본 보존)
gims_df = gims_df_orig.copy()

# ==============================================================================
# 3. 법적 기준 기반 '오염 이벤트' 정의 (수동 매핑으로 안정성 확보)
# ==============================================================================
print("\n■■■ 3. 법적 기준 기반 '오염 이벤트' 정의 시작 ■■■")
drinking_standards = standards_df[standards_df['구분'] == '음용수'].copy()
def parse_std_value(value):
    if isinstance(value, str):
        value = value.strip()
        if '<=' in value: return float(value.replace('<=',''))
        if '-' in value: return tuple(map(float, value.split('-')))
        if '불검출' in value: return 0.0
    return np.nan
drinking_standards['processed_std'] = drinking_standards['기준값'].apply(parse_std_value)
drinking_standards.dropna(subset=['processed_std'], inplace=True)
final_mapper_dict = {
    '암모니아성질소_NH3_N': '암모니아성질소', '질산염': '질산성질소', '비소_As': '비소', '수은_Hg': '수은',
    '카드뮴_Cd': '카드뮴', '납_Pb': '납', '크롬': '크롬', '벤젠_Benzene': '벤젠', '톨루엔_Toluene': '톨루엔',
    '에틸벤젠_Ethylbenzene': '에틸벤젠', '크실렌_Xylene': '크실렌', '페놀_Phenol': '페놀',
    '트리클로로에틸렌_TCE': '트리클로로에틸렌', '테트라클로로에틸렌': '테트라클로로에틸렌',
    '사염화탄소': '사염화탄소', '디클로로메탄_Dichlomethans': '디클로로메탄', '1_4_다이옥산': '1,4-다이옥산',
    '페니트로티온': '페니트로티온', '파라티온_Parathion': '파라티온', '카바릴_Carbaryl': '카바릴',
    '다이아지논_Diazion': '다이아지논', '일반세균': '일반세균', '총대장균군': '총대장균군',
    '분원성대장균군': '분원성대장균군', '시안_CN': '시안', '불소_F': '불소', '대장균군': '대장균', '세레늄_Se': '셀레늄'}
valid_mapper = {gims_col: std_name for gims_col, std_name in final_mapper_dict.items() if gims_col in gims_df.columns}
mapper_df = pd.DataFrame(list(valid_mapper.items()), columns=['gims_col', 'std_name'])
final_standards = pd.merge(drinking_standards, mapper_df, left_on='항목', right_on='std_name')
exceed_cols = []
for _, row in final_standards.iterrows():
    gims_col = row['gims_col']; std_val = row['processed_std']; exceed_col = f"{gims_col}_기준초과"; exceed_cols.append(exceed_col)
    gims_df[gims_col] = pd.to_numeric(gims_df[gims_col], errors='coerce')
    if isinstance(std_val, tuple): gims_df[exceed_col] = (gims_df[gims_col] < std_val[0]) | (gims_df[gims_col] > std_val[1])
    else: gims_df[exceed_col] = gims_df[gims_col] > std_val
gims_df['총_기준초과건수'] = gims_df[exceed_cols].sum(axis=1)
print(f"  ✅ 총 {len(final_standards)}개 항목에 대한 법적 기준 적용 완료.")


# ==============================================================================
# 4. [상세 분석 및 시각화 ①] 오염 현황 상세 분석 (The WHERE)
# ==============================================================================
print("\n■■■ 4. [상세 분석] 법적 기준 초과 현황 분석 시작 ■■■")
total_exceed_events = gims_df['총_기준초과건수'].sum()
print(f"\n  ▶ 총 발견된 개별 오염 이벤트(법적 기준 초과) 건수: {int(total_exceed_events)} 건")

# 어떤 물질이 가장 문제인가?
exceed_by_item = gims_df[exceed_cols].sum().sort_values(ascending=False)
exceed_by_item = exceed_by_item[exceed_by_item > 0] # 0건인 항목은 제외
print("\n--- [물질별] 기준 초과 건수 Top 10 ---")
print(exceed_by_item.head(10))

# [시각화 ①] 물질별 기준 초과 건수
plt.figure(figsize=(12, 8))
ax = sns.barplot(x=exceed_by_item.head(10).values, y=exceed_by_item.head(10).index, palette='viridis')
ax.set_title('가장 빈번하게 기준을 초과한 오염 물질 Top 10', fontsize=18, weight='bold')
ax.set_xlabel('총 초과 건수 (건)', fontsize=12)
ax.set_ylabel('')
# 막대 위에 숫자 표시
for p in ax.patches:
    ax.annotate(f'{int(p.get_width())} 건', (p.get_width(), p.get_y() + p.get_height() / 2.), 
                ha='left', va='center', xytext=(5, 0), textcoords='offset points', fontsize=12)
plt.show()

# 어떤 관측소(공번호)에서 가장 많이 발생했는가?
exceed_by_well = gims_df.groupby('공번호')['총_기준초과건수'].sum().sort_values(ascending=False)
exceed_by_well = exceed_by_well[exceed_by_well > 0]
print("\n--- [관측소별] 기준 초과 건수 Top 10 ---")
print(exceed_by_well.head(10))

# [시각화 ②] 관측소별 기준 초과 건수
plt.figure(figsize=(12, 8))
ax = sns.barplot(x=exceed_by_well.head(10).values, y=exceed_by_well.head(10).index, palette='plasma')
ax.set_title('가장 빈번하게 기준을 초과한 관측소 Top 10', fontsize=18, weight='bold')
ax.set_xlabel('총 초과 건수 (건)', fontsize=12)
ax.set_ylabel('관측소 번호(공번호)')
for p in ax.patches:
    ax.annotate(f'{int(p.get_width())} 건', (p.get_width(), p.get_y() + p.get_height() / 2.), 
                ha='left', va='center', xytext=(5, 0), textcoords='offset points', fontsize=12)
plt.show()

# ==============================================================================
# 5. 데이터 병합을 위한 '지역명' 키 생성 및 표준화
# ==============================================================================
print("\n■■■ 5. 데이터 병합을 위한 키 생성 및 표준화 시작 ■■■")
def standardize_region_name(sido, sigungu):
    if pd.isna(sido) or pd.isna(sigungu): return None
    sido = str(sido).replace('특별자치도', '').replace('광역시', '').replace('특별시', '').replace('특별자치시', '').strip()
    sigungu = str(sigungu).strip()
    return f"{sido}_{sigungu}"
    
def standardize_feature_name(name):
     if pd.isna(name): return None
     name = name.replace('강원특별자치도','강원'); name = re.sub(r'\s+', '', name)
     return name

gims_df['지역명_std'] = gims_df.apply(lambda row: standardize_region_name(row['시군구명'].split(" ")[0], row['시군구명'].split(" ")[1] if len(row['시군구명'].split(" ")) > 1 else ""), axis=1)
feature_df.rename(columns={'지역': '지역명'}, inplace=True)
feature_df['지역명_std'] = feature_df['지역명'].apply(standardize_feature_name)
print("  ✅ 양쪽 데이터의 지역명 키 생성 및 표준화 완료.")

# ==============================================================================
# 6. [상세 분석 및 시각화 ②] The WHO 분석
# ==============================================================================
print("\n■■■ 6. 'The WHO' 분석 시작 ■■■")
sigungu_gims_agg = gims_df.groupby('지역명_std')['총_기준초과건수'].sum().reset_index()
merged_df = pd.merge(sigungu_gims_agg, feature_df, on='지역명_std', how='inner')
print(f"  ✅ GIMS-Feature Matrix 병합 완료. (공통 지역 {len(merged_df)}개)")

if not merged_df.empty and merged_df['총_기준초과건수'].sum() > 0:
    source_cols = [col for col in feature_df.select_dtypes(include=np.number) if '_std' not in str(col)]
    correlation = merged_df[['총_기준초과건수'] + source_cols].corr(numeric_only=True)
    corr_with_target = correlation['총_기준초과건수'].drop('총_기준초과건수').sort_values(ascending=False)
    
    # [시각화 ③] 오염 발생과 오염원 특성 간의 상관관계
    plt.figure(figsize=(12, 8))
    # 양수/음수 구분을 위해 diverging palette 'coolwarm' 사용
    ax = sns.barplot(x=corr_with_target.values, y=corr_with_target.index, palette='coolwarm')
    ax.set_title('총 오염 발생 건수와 오염원 특성 간의 상관관계', fontsize=18, weight='bold')
    ax.set_xlabel('상관 계수 (Pearson)', fontsize=12)
    ax.set_ylabel('오염원 특성')
    ax.axvline(x=0, color='black', linewidth=0.8, linestyle='--')
    for p in ax.patches:
        value = p.get_width()
        x = p.get_width()
        y = p.get_y() + p.get_height() / 2
        # 소수점 3자리까지 표시
        ax.annotate(f'{value:.3f}', (x, y), ha='left' if value >= 0 else 'right', va='center', 
                    xytext=(5 if value >= 0 else -5, 0), textcoords='offset points')
    plt.show()

else:
    print("\n[알림] 데이터 병합에 실패했거나 유효 오염 건수가 없어 상관관계 분석을 건너뜁니다.")

print("\n■■■ 최종 분석 완료 ■■■")
```

```python
# ==============================================================================
# 0. 환경 설정 및 모든 라이브러리 임포트
# ==============================================================================
print("■■■ 0. 환경 설정 및 라이브러리 임포트 시작 ■■■")
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import seaborn as sns
import warnings
import re

warnings.filterwarnings('ignore')

# ==============================================================================
# 1. 안정적인 한글 폰트 설정 및 전문가용 시각화 테마 적용
# ==============================================================================
print("\n■■■ 1. 한글 폰트 및 시각화 테마 설정 시작 ■■■")
try:
    font_path = "c:/Windows/Fonts/malgun.ttf"
    font_name = fm.FontProperties(fname=font_path).get_name()
    # [업그레이드] Seaborn 테마를 설정하고, 기본 폰트를 지정합니다.
    sns.set_theme(style='whitegrid', font=font_name, palette='viridis')
    plt.rcParams['axes.unicode_minus'] = False
    print("  ✅ 폰트 및 전문가용 시각화 테마 적용 완료.")
except Exception as e:
    print(f"  - 폰트 설정 중 오류 발생: {e}")

# ==============================================================================
# 2. 데이터 로드 및 사전 작업
# ==============================================================================
print("\n■■■ 2. 데이터 로드 및 핵심 사전 작업 시작 ■■■")

def load_csv(path):
    try: return pd.read_csv(path, encoding='cp949', low_memory=False)
    except UnicodeDecodeError: return pd.read_csv(path, encoding='utf-8', low_memory=False)

gims_df_orig = load_csv('./data/gims_master_data2.csv')
feature_df = load_csv('./data/feature_matrix_by_sigungu.csv')
standards_df = load_csv('./data/groundwater_quality_standards.csv')

if any(df is None for df in [gims_df_orig, feature_df, standards_df]):
    print("\n‼️ 일부 파일 로드에 실패하여 분석을 중단합니다."); exit()
print("  ✅ 모든 원본 파일 로드 성공.")
gims_df = gims_df_orig.copy()

# ==============================================================================
# 3. 법적 기준 기반 '오염 이벤트' 정의 (수동 매핑으로 안정성 확보)
# ==============================================================================
print("\n■■■ 3. 법적 기준 기반 '오염 이벤트' 정의 시작 ■■■")
drinking_standards = standards_df[standards_df['구분'] == '음용수'].copy()
def parse_std_value(value):
    if isinstance(value, str):
        value = value.strip()
        if '<=' in value: return float(value.replace('<=',''))
        if '-' in value: return tuple(map(float, value.split('-')))
        if '불검출' in value: return 0.0
    return np.nan
drinking_standards['processed_std'] = drinking_standards['기준값'].apply(parse_std_value)
drinking_standards.dropna(subset=['processed_std'], inplace=True)
final_mapper_dict = {
    '암모니아성질소_NH3_N': '암모니아성질소', '질산염': '질산성질소', '비소_As': '비소', '수은_Hg': '수은', '카드뮴_Cd': '카드뮴', '납_Pb': '납', '크롬': '크롬',
    '벤젠_Benzene': '벤젠', '톨루엔_Toluene': '톨루엔', '에틸벤젠_Ethylbenzene': '에틸벤젠', '크실렌_Xylene': '크실렌', '페놀_Phenol': '페놀',
    '트리클로로에틸렌_TCE': '트리클로로에틸렌', '테트라클로로에틸렌': '테트라클로로에틸렌', '사염화탄소': '사염화탄소', '디클로로메탄_Dichlomethans': '디클로로메탄',
    '1_4_다이옥산': '1,4-다이옥산', '페니트로티온': '페니트로티온', '파라티온_Parathion': '파라티온', '카바릴_Carbaryl': '카바릴', '다이아지논_Diazion': '다이아지논',
    '일반세균': '일반세균', '총대장균군': '총대장균군', '분원성대장균군': '분원성대장균군', '시안_CN': '시안', '불소_F': '불소', '대장균군': '대장균', '세레늄_Se': '셀레늄'}
valid_mapper = {gims_col: std_name for gims_col, std_name in final_mapper_dict.items() if gims_col in gims_df.columns}
mapper_df = pd.DataFrame(list(valid_mapper.items()), columns=['gims_col', 'std_name'])
final_standards = pd.merge(drinking_standards, mapper_df, left_on='항목', right_on='std_name')
exceed_cols = []
for _, row in final_standards.iterrows():
    gims_col = row['gims_col']; std_val = row['processed_std']; exceed_col = f"{gims_col}_기준초과"; exceed_cols.append(exceed_col)
    gims_df[gims_col] = pd.to_numeric(gims_df[gims_col], errors='coerce')
    if isinstance(std_val, tuple): gims_df[exceed_col] = (gims_df[gims_col] < std_val[0]) | (gims_df[gims_col] > std_val[1])
    else: gims_df[exceed_col] = gims_df[gims_col] > std_val
gims_df['총_기준초과건수'] = gims_df[exceed_cols].sum(axis=1)
print(f"  ✅ 총 {len(final_standards)}개 항목에 대한 법적 기준 적용 완료.")


# ==============================================================================
# 4. [상세 분석 및 시각화 ①] 오염 현황 상세 분석 (The WHERE)
# ==============================================================================
print("\n■■■ 4. [상세 분석] 법적 기준 초과 현황 분석 시작 ■■■")
total_exceed_events = gims_df['총_기준초과건수'].sum()
print(f"\n  ▶ 총 발견된 개별 오염 이벤트(법적 기준 초과) 건수: {int(total_exceed_events)} 건")

exceed_by_item = gims_df[exceed_cols].sum().sort_values(ascending=False)
exceed_by_item = exceed_by_item[exceed_by_item > 0]
print("\n--- [물질별] 기준 초과 건수 (전체) ---")
print(exceed_by_item)
plt.figure(figsize=(12, 10)); ax = sns.barplot(x=exceed_by_item.values, y=exceed_by_item.index, palette='viridis')
ax.set_title('물질별 기준 초과 건수', fontsize=18, weight='bold'); ax.set_xlabel('총 초과 건수 (건)'); ax.set_ylabel('')
for p in ax.patches: ax.annotate(f'{int(p.get_width())}', (p.get_width(), p.get_y()+p.get_height()/2.),ha='left',va='center',xytext=(5,0),textcoords='offset points')
plt.show()

exceed_by_well = gims_df.groupby('공번호')['총_기준초과건수'].sum().sort_values(ascending=False)
exceed_by_well = exceed_by_well[exceed_by_well > 0]
print("\n--- [관측소별] 기준 초과 건수 (전체) ---")
print(exceed_by_well)
plt.figure(figsize=(12, 12)); ax = sns.barplot(x=exceed_by_well.head(20).values, y=exceed_by_well.head(20).index, palette='plasma')
ax.set_title('기준 초과 건수 상위 20개 관측소', fontsize=18, weight='bold'); ax.set_xlabel('총 초과 건수 (건)'); ax.set_ylabel('관측소 번호(공번호)')
for p in ax.patches: ax.annotate(f'{int(p.get_width())}',(p.get_width(),p.get_y()+p.get_height()/2.),ha='left',va='center',xytext=(5,0),textcoords='offset points')
plt.show()


# ==============================================================================
# 5. 데이터 병합을 위한 '지역명' 키 생성 및 표준화
# ==============================================================================
print("\n■■■ 5. 데이터 병합을 위한 키 생성 및 표준화 시작 ■■■")
def standardize_region_name(sido, sigungu):
    if pd.isna(sido) or pd.isna(sigungu): return None
    sido = str(sido).replace('특별자치도', '').replace('광역시', '').replace('특별시', '').replace('특별자치시', '').strip()
    sigungu = str(sigungu).strip(); return f"{sido}_{sigungu}"
def standardize_feature_name(name):
     if pd.isna(name): return None
     name = name.replace('강원특별자치도','강원'); name = re.sub(r'\s+', '', name)
     name = name.replace('_', ' ') # feature_df는 _ 대신 공백을 기준으로 분리될 수 있음
     parts = name.split(" ")
     if len(parts) > 1: return standardize_region_name(parts[0], "".join(parts[1:]))
     return name

gims_df['지역명_std'] = gims_df.apply(lambda row: standardize_region_name(row['시군구명'].split(" ")[0], row['시군구명'].split(" ")[1] if len(row['시군구명'].split(" ")) > 1 else ""), axis=1)
feature_df.rename(columns={'지역': '지역명'}, inplace=True)
feature_df['지역명_std'] = feature_df['지역명'].apply(standardize_feature_name)
print("  ✅ 양쪽 데이터의 지역명 키 생성 및 표준화 완료.")

# ==============================================================================
# 6. [상세 분석 및 시각화 ②] The WHO 분석
# ==============================================================================
print("\n■■■ 6. 'The WHO' 분석 시작 ■■■")
sigungu_gims_agg = gims_df.groupby('지역명_std')['총_기준초과건수'].sum().reset_index()
merged_df = pd.merge(sigungu_gims_agg, feature_df, on='지역명_std', how='inner')
print(f"  ✅ GIMS-Feature Matrix 병합 완료. (공통 지역 {len(merged_df)}개)")

if not merged_df.empty and merged_df['총_기준초과건수'].sum() > 0:
    source_cols = [col for col in feature_df.select_dtypes(include=np.number) if '_std' not in str(col)]
    correlation = merged_df[['총_기준초과건수'] + source_cols].corr(numeric_only=True)
    corr_with_target = correlation['총_기준초과건수'].drop('총_기준초과건수').sort_values(ascending=False)
    plt.figure(figsize=(12, 8)); ax = sns.barplot(x=corr_with_target.values, y=corr_with_target.index, palette='coolwarm')
    ax.set_title('총 오염 발생 건수와 오염원 특성 간의 상관관계', fontsize=18, weight='bold')
    ax.set_xlabel('상관 계수 (Pearson)'); ax.set_ylabel('오염원 특성'); ax.axvline(x=0, color='black', linewidth=0.8, linestyle='--')
    for p in ax.patches:
        value = p.get_width(); x = p.get_width(); y = p.get_y() + p.get_height() / 2
        ax.annotate(f'{value:.3f}', (x, y), ha='left' if value >= 0 else 'right', va='center', xytext=(5 if value>=0 else -5,0), textcoords='offset points')
    plt.show()
else:
    print("\n[알림] 데이터 병합에 실패했거나 유효 오염 건수가 없어 상관관계 분석을 건너뜁니다.")

print("\n■■■ 최종 분석 완료 ■■■")
```

```python
# ==============================================================================
# 0. 환경 설정 및 모든 라이브러리 임포트
# ==============================================================================
print("■■■ 0. 환경 설정 및 라이브러리 임포트 시작 ■■■")
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import seaborn as sns
import warnings
import re
from difflib import SequenceMatcher # [핵심] 문자열 유사도 비교를 위한 라이브러리

warnings.filterwarnings('ignore')

# ==============================================================================
# 1. 안정적인 한글 폰트 설정 및 전문가용 시각화 테마 적용
# ==============================================================================
print("\n■■■ 1. 한글 폰트 및 시각화 테마 설정 시작 ■■■")
try:
    font_path = "c:/Windows/Fonts/malgun.ttf"
    font_name = fm.FontProperties(fname=font_path).get_name()
    sns.set_theme(style='whitegrid', font=font_name, palette='viridis')
    plt.rcParams['axes.unicode_minus'] = False
    print("  ✅ 폰트 및 전문가용 시각화 테마 적용 완료.")
except Exception as e:
    print(f"  - 폰트 설정 중 오류 발생: {e}")

# ==============================================================================
# 2. 데이터 로드 및 사전 작업
# ==============================================================================
print("\n■■■ 2. 데이터 로드 및 핵심 사전 작업 시작 ■■■")
def load_csv(path):
    try: return pd.read_csv(path, encoding='cp949', low_memory=False)
    except UnicodeDecodeError: return pd.read_csv(path, encoding='utf-8', low_memory=False)

gims_df_orig = load_csv('./data/gims_master_data2.csv')
feature_df = load_csv('./data/feature_matrix_by_sigungu.csv')
standards_df = load_csv('./data/groundwater_quality_standards.csv')

if any(df is None for df in [gims_df_orig, feature_df, standards_df]):
    print("\n‼️ 일부 파일 로드에 실패하여 분석을 중단합니다."); exit()
print("  ✅ 모든 원본 파일 로드 성공.")
gims_df = gims_df_orig.copy()

# ==============================================================================
# 3. 법적 기준 기반 '오염 이벤트' 정의 (지능형 자동 매핑)
# ==============================================================================
print("\n■■■ 3. 법적 기준 기반 '오염 이벤트' 정의 시작 (지능형 자동 매핑) ■■■")
drinking_standards = standards_df[standards_df['구분'] == '음용수'].copy()
def parse_std_value(value):
    if isinstance(value, str):
        value = value.strip()
        if '<=' in value: return float(value.replace('<=',''))
        if '-' in value: return tuple(map(float, value.split('-')))
        if '불검출' in value: return 0.0
    return np.nan
drinking_standards['processed_std'] = drinking_standards['기준값'].apply(parse_std_value)
drinking_standards.dropna(subset=['processed_std'], inplace=True)

# --- [핵심 업그레이드] AI가 스스로 판단하는 지능형 매핑 로직 ---
print("\n  ▶ gims-기준치 물질명 지능형 자동 매핑 및 결과 출력...")
def get_best_match(gims_col, std_list, min_similarity=0.7):
    best_match = None; highest_sim = min_similarity
    def clean(name): return re.sub(r'\(.*\)|_[A-Za-z0-9]+|\s', '', name).replace('성', '')
    clean_gims = clean(gims_col)
    for std_item in std_list:
        similarity = SequenceMatcher(None, clean(std_item), clean_gims).ratio()
        if '질산염' in clean_gims and '질산' in clean(std_item): similarity = 1.0
        if similarity > highest_sim: highest_sim = similarity; best_match = std_item
    return best_match, highest_sim

std_items_available = drinking_standards['항목'].unique().tolist(); mapped_pairs = []
print("    [상세 매핑 결과]")
for gims_col in gims_df.columns:
    best_match_std, similarity = get_best_match(gims_col, std_items_available)
    if best_match_std:
        mapped_pairs.append({'gims_col': gims_col, 'std_name': best_match_std})
        std_items_available.remove(best_match_std) # 중복 매칭 방지
        print(f"      - GIMS:'{gims_col}' <---> 기준:'{best_match_std}' (유사도: {similarity:.2f})")

mapper_df = pd.DataFrame(mapped_pairs)
final_standards = pd.merge(drinking_standards, mapper_df, left_on='항목', right_on='std_name')
print(f"    ✅ 자동 매핑 완료: 총 {len(final_standards)}개 항목에 대한 법적 기준 확보.")
# --- (이하 코드는 동일) ---
exceed_cols = []
for _, row in final_standards.iterrows():
    gims_col = row['gims_col']; std_val = row['processed_std']; exceed_col = f"{gims_col}_기준초과"; exceed_cols.append(exceed_col)
    gims_df[gims_col] = pd.to_numeric(gims_df[gims_col], errors='coerce')
    if isinstance(std_val, tuple): gims_df[exceed_col] = (gims_df[gims_col] < std_val[0]) | (gims_df[gims_col] > std_val[1])
    else: gims_df[exceed_col] = gims_df[gims_col] > std_val
gims_df['총_기준초과건수'] = gims_df[exceed_cols].sum(axis=1)


# ==============================================================================
# 4. [상세 분석 및 시각화] 오염 현황 상세 분석 (The WHERE)
# ==============================================================================
print("\n■■■ 4. [상세 분석] 법적 기준 초과 현황 분석 시작 ■■■")
total_exceed_events = gims_df['총_기준초과건수'].sum()
print(f"\n  ▶ 총 발견된 개별 오염 이벤트(법적 기준 초과) 건수: {int(total_exceed_events)} 건")

# 어떤 물질이 가장 문제인가? (Top 10 제한 해제)
exceed_by_item = gims_df[exceed_cols].sum().sort_values(ascending=False)
exceed_by_item = exceed_by_item[exceed_by_item > 0] # 0건인 항목은 제외
print("\n--- [물질별] 기준 초과 건수 (전체) ---")
print(exceed_by_item)
plt.figure(figsize=(12, 10)); ax = sns.barplot(x=exceed_by_item.values, y=exceed_by_item.index)
ax.set_title('물질별 기준 초과 건수', fontsize=18, weight='bold'); ax.set_xlabel('총 초과 건수 (건)'); ax.set_ylabel('')
for p in ax.patches: ax.annotate(f'{int(p.get_width())}', (p.get_width(), p.get_y()+p.get_height()/2.),ha='left',va='center',xytext=(5,0),textcoords='offset points')
plt.show()

# 어떤 관측소(공번호)에서 가장 많이 발생했는가? (상위 20개만 시각화)
exceed_by_well = gims_df.groupby('공번호')['총_기준초과건수'].sum().sort_values(ascending=False)
exceed_by_well = exceed_by_well[exceed_by_well > 0]
print("\n--- [관측소별] 기준 초과 건수 (전체) ---")
print(exceed_by_well)
plt.figure(figsize=(12, 12)); ax = sns.barplot(x=exceed_by_well.head(20).values, y=exceed_by_well.head(20).index, palette='plasma')
ax.set_title('기준 초과 건수 상위 20개 관측소', fontsize=18, weight='bold'); ax.set_xlabel('총 초과 건수 (건)'); ax.set_ylabel('관측소 번호(공번호)')
for p in ax.patches: ax.annotate(f'{int(p.get_width())}',(p.get_width(),p.get_y()+p.get_height()/2.),ha='left',va='center',xytext=(5,0),textcoords='offset points')
plt.show()

# ==============================================================================
# 5. 데이터 병합을 위한 '지역명' 키 생성 및 표준화
# ==============================================================================
print("\n■■■ 5. 데이터 병합을 위한 키 생성 및 표준화 시작 ■■■")
def standardize_region_name(sido, sigungu):
    if pd.isna(sido) or pd.isna(sigungu): return None
    sido = str(sido).replace('특별자치도', '').replace('광역시', '').replace('특별시', '').replace('특별자치시', '').strip()
    sigungu = str(sigungu).strip(); return f"{sido}_{sigungu}"
def standardize_feature_name(name):
     if pd.isna(name): return None
     name = name.replace('강원특별자치도','강원'); name = re.sub(r'\s+', '', name)
     name = name.replace('_', ' ') # feature_df는 _ 대신 공백을 기준으로 분리될 수 있음
     parts = name.split(" ")
     if len(parts) > 1: return standardize_region_name(parts[0], "".join(parts[1:]))
     return name

gims_df['지역명_std'] = gims_df.apply(lambda row: standardize_region_name(row['시군구명'].split(" ")[0], row['시군구명'].split(" ")[1] if len(row['시군구명'].split(" ")) > 1 else ""), axis=1)
feature_df.rename(columns={'지역': '지역명'}, inplace=True)
feature_df['지역명_std'] = feature_df['지역명'].apply(standardize_feature_name)
print("  ✅ 양쪽 데이터의 지역명 키 생성 및 표준화 완료.")

# ==============================================================================
# 6. 'The WHO' 분석 (상관관계 시각화)
# ==============================================================================
print("\n■■■ 6. 'The WHO' 분석 시작 ■■■")
sigungu_gims_agg = gims_df.groupby('지역명_std')['총_기준초과건수'].sum().reset_index()
merged_df = pd.merge(sigungu_gims_agg, feature_df, on='지역명_std', how='inner')
print(f"  ✅ GIMS-Feature Matrix 병합 완료. (공통 지역 {len(merged_df)}개)")

if not merged_df.empty and merged_df['총_기준초과건수'].sum() > 0:
    source_cols = [col for col in feature_df.select_dtypes(include=np.number) if '_std' not in str(col)]
    correlation = merged_df[['총_기준초과건수'] + source_cols].corr(numeric_only=True)
    corr_with_target = correlation['총_기준초과건수'].drop('총_기준초과건수').sort_values(ascending=False)
    plt.figure(figsize=(12, 8)); ax = sns.barplot(x=corr_with_target.values, y=corr_with_target.index, palette='coolwarm')
    ax.set_title('총 오염 발생 건수와 오염원 특성 간의 상관관계', fontsize=18, weight='bold')
    ax.set_xlabel('상관 계수 (Pearson)'); ax.set_ylabel('오염원 특성'); ax.axvline(x=0, color='black', linewidth=0.8, linestyle='--')
    for p in ax.patches:
        value = p.get_width(); x = p.get_width(); y = p.get_y() + p.get_height() / 2
        ax.annotate(f'{value:.3f}', (x, y), ha='left' if value >= 0 else 'right', va='center', xytext=(5 if value>=0 else -5,0), textcoords='offset points')
    plt.show()
else:
    print("\n[알림] 데이터 병합에 실패했거나 유효 오염 건수가 없어 상관관계 분석을 건너뜁니다.")

print("\n■■■ 최종 분석 완료 ■■■")
```
