
### **2단계 :  데이터 처리 및 통합 (Core Dataset 생성)**

```python
import pandas as pd

# --- 데이터 처리 로직 시작 ---

# 1. 필요한 컬럼만 선택
# 수질 데이터에서 필요한 컬럼 목록
cols_quality = [
    '공번호',
    '조사시기',
    # --- 질소/인 화합물 (영양염류) ---
    'NH4_N',
    '암모니아성질소_NH3_N',
    '유기인',
    '인산염인',
    '아질산이온',
    '질산염',

    # --- 중금속 및 금속류 ---
    '비소_As',
    '수은_Hg',
    '카드뮴_Cd',
    '납_Pb',
    '6가크롬_Cr+6',
    '크롬',
    '구리_Cu',
    '아연_Zn',
    '망간_Mn',
    '철_Fe',
    '알루미늄_Al',
    '세레늄_Se',
    '니켈',
    '바륨',

    # --- 휘발성 유기화합물(VOCs) 및 유기물질 ---
    '벤젠_Benzene',
    '톨루엔_Toluene',
    '에틸벤젠_Ethylbenzene',
    '크실렌_Xylene',
    '페놀_Phenol',
    '트리클로로에틸렌_TCE',
    '테트라클로로에틸렌',
    '사염화탄소',
    '디클로로메탄_Dichlomethans',
    '총트리할로메탄_THM',
    '1_4_다이옥산',

    # --- 농약류 ---
    # '유기인'은 중복되므로 한 번만 포함
    '페니트로티온',
    '파라티온_Parathion',
    '카바릴_Carbaryl',
    '말라티온_Malathion',
    '다이아지논_Diazion',

    # --- 미생물 ---
    '일반세균',
    '일반세균_저온',
    '일반세균_중온',
    '총대장균군',
    '대장균군',
    '분원성대장균군',
    '녹농균',
    '살모넬라',

    # --- 기타 이온 및 일반 오염 지표 ---
    '시안_CN',
    '불소_F',
    '염소이온_Cl_',
    '황산이온_SO4_2',
    '세제_ABS',
    '화학적산소요구량_COD',
]

# 수리상수 데이터에서 필요한 컬럼 목록
cols_hydro = [
    '공번호',
    '시도CODE',
    '시군구CODE',
    'K_m_day',     # 수리전도도
    'T_m2_day',    # 투수량계수
    '저류계수',

    'TM_X_5186',   # X좌표
    'TM_Y_5186'    # Y좌표
]

# 컬럼 목록을 기반으로 데이터프레임 필터링
df_quality_selected = ddf[cols_quality]
df_hydro_selected = df[cols_hydro]

# 수리상수 데이터는 '공번호' 기준으로 중복된 행이 있을 수 있으므로, 최신 데이터만 남기기 위해 중복을 제거합니다.
df_hydro_unique = df_hydro_selected.drop_duplicates(subset=['공번호'], keep='last').reset_index(drop=True)
print(f"  - 수리상수 데이터 중복 제거 완료: {len(df_hydro_selected)} -> {len(df_hydro_unique)}")

# 2. '공번호'를 Key로 사용하여 두 데이터프레임 병합
print("\n2. '공번호'를 기준으로 두 데이터를 병합합니다 (pd.merge)...")
# 수질 데이터(df_quality_selected)는 시계열이므로 그대로 두고,
# 관측정별로 고유한 값을 가지는 수리상수 데이터(df_hydro_unique)를 병합합니다.
# how='inner'는 양쪽 데이터에 모두 '공번호'가 존재하는 경우만 남깁니다.
gims_master_data = pd.merge(df_quality_selected, df_hydro_unique, on='공번호', how='inner')

# 3. 결과 확인 및 저장
print("\n3. 최종 통합 데이터 'gims_master_data.csv'를 생성합니다...")
# 최종 결과물의 상위 5개 행 출력
print("\n[병합 결과 미리보기 (상위 5개)]")
print(gims_master_data.head())

# 최종 결과물의 정보 요약 출력
print("\n[최종 데이터 정보]")
gims_master_data.info()

# 코랩 환경에 CSV 파일로 저장
# encoding='utf-8-sig'는 Excel에서 한글이 깨지지 않게 하기 위함입니다.
output_filename = './data/gims_master_data.csv'
gims_master_data.to_csv(output_filename, index=False, encoding='utf-8-sig')

print(f"\n작업 완료! '{output_filename}' 파일이 현재 디렉토리에 저장되었습니다.")

```

### GIMS 데이터 시군구CODE - 시군구명(법정동코드) 맵핑

```python


# 2. pandas import 및 경로 설정
import pandas as pd

# 파일 경로
gims_path = './data/gims_master_data.csv'
lawcode_path = './data/법정동시군구/법정동코드 전체자료.csv'

# 3. 파일 불러오기
df_gims = pd.read_csv(gims_path)
df_law = pd.read_csv(lawcode_path, encoding='utf-8-sig')

# 4. GIMS 코드 전처리: 문자열로 변환 후 앞 10자리 추출
df_gims['시군구코드_10자리'] = df_gims['시군구CODE'].astype(str).str[:10]

# 5. 병합을 위한 컬럼 이름 통일
df_law = df_law.rename(columns={'법정동코드': '시군구코드_10자리', '법정동명': '시군구명'})
df_law['시군구코드_10자리'] = df_law['시군구코드_10자리'].astype(str)

# 6. 병합 수행 (left join)
df_merged = df_gims.merge(df_law[['시군구코드_10자리', '시군구명']], on='시군구코드_10자리', how='left')

# 7. 결과 확인
print("✅병합 완료! 병합된 데이터 예시:")
print(df_merged[['시군구CODE', '시군구명']].head())

# 8. 저장 경로 및 저장
save_path = './data/gims_master_data2.csv'
df_merged.to_csv(save_path, index=False, encoding='utf-8-sig')

print(" 저장 완료! 경로:")
print(save_path)
```
